```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

house_file_path = '/content/sample_data/tvmarketing.csv'


tv_data = pd.read_csv(house_file_path)


feature_names = ['TV']
X = tv_data[feature_names]


target_name = ['Sales']
y = tv_data[target_name]


#Visualizamos los datos en un grafico de dispersion

#podemos ver de primeras que los datos no estan tan dispersos.

plt.scatter(X,y)
plt.xlabel("Tv")
plt.ylabel("Sales")

plt.show()



X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3)



linear_regression = linear_model.LinearRegression()

linear_regression.fit(X_train, y_train)

y_precio_tv_pred = linear_regression.predict(X_test)

# The coefficients
print("Coefficients: \n", linear_regression.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_precio_tv_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_precio_tv_pred))


# Finalmente mostramos el grafico final.
plt.scatter(X_test, y_test, color="black")
plt.plot(X_test, y_precio_tv_pred, color="blue", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()



#Mostramos la diferencia con el resultado real:

# Actual vs Predicted

#ponemos 41 porque es el tamaño de y_test (siempre hay que tener esto claro.)

c = [i for i in range(1,61,1)]         # generating index
fig = plt.figure()
plt.plot(c,y_test, color="blue", linewidth=2, linestyle="-")
plt.plot(c,y_precio_tv_pred, color="red",  linewidth=2, linestyle="-")
fig.suptitle('Actual and Predicted', fontsize=20)              # Plot heading
plt.xlabel('Index', fontsize=18)                               # X-label
plt.ylabel('Sales', fontsize=16)
```
